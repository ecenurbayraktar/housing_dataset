```python
train_dataset=train_dataset.drop("PID",axis=1)
test_dataset=test_dataset.drop("PID",axis=1)
```

```python
#pool gc sütunundaki NaN değeri eksik veriden değil de havuz olmadığı için

train_dataset["Pool QC"]=train_dataset["Pool QC"].fillna("none")
train_dataset["Pool QC"].value_counts(dropna=False)
test_dataset["Pool QC"]=test_dataset["Pool QC"].fillna("none")
test_dataset["Pool QC"].value_counts(dropna=False)
```

```python
print(type(train_dataset))
```

```python
none_columns = [
    "Alley",
    "Fence",
    "Fireplace Qu",
    "Garage Type",
    "Garage Finish",
    "Garage Qual",
    "Garage Cond",
    "Bsmt Qual",
    "Bsmt Cond",
    "Bsmt Exposure",
    "BsmtFin Type 1",
    "BsmtFin Type 2",
    "Misc Feature"
]

for col in none_columns:
    train_dataset[col] = train_dataset[col].fillna("None")
    test_dataset[col] = test_dataset[col].fillna("None")
```

```python
has_feature_columns = {
    "Pool QC": "new_HasPool",
    "Garage Type": "new_HasGarage",
    "Fireplace Qu": "new_HasFireplace",
    "Fence": "new_HasFence",
    "Alley": "new_HasAlley",
    "Bsmt Qual": "new_HasBasement",
    "Misc Feature": "new_HasMiscFeature"
}

for original_col, new_col in has_feature_columns.items():
    train_dataset[new_col] = (train_dataset[original_col] != "None").astype(int)
    test_dataset[new_col] = (test_dataset[original_col] != "None").astype(int)
```

```python
train_dataset["new_IsNew"] = train_dataset["Year Built"].apply(lambda x: "New" if x >= 2000 else "Old")
train_dataset["new_GarageSize"] = train_dataset["Garage Area"].apply(lambda x: "Large" if x > 500.000000 else "Small")
test_dataset["new_IsNew"] = test_dataset["Year Built"].apply(lambda x: "New" if x >= 2000 else "Old")
test_dataset["new_GarageSize"] = test_dataset["Garage Area"].apply(lambda x: "Large" if x > 500.000000 else "Small")
```

```python
def check_garage_built_later(row):
    if pd.isna(row["Garage Yr Blt"]):
        return "No Garage"
    elif row["Garage Yr Blt"] > row["Year Built"]:
        return "New"
    else:
        return "Old"

train_dataset["new_IsGarageNew"] = train_dataset.apply(check_garage_built_later, axis=1)
test_dataset["new_IsGarageNew"] = test_dataset.apply(check_garage_built_later, axis=1)

print(train_dataset["new_IsGarageNew"].value_counts())
```

```python
def check_house_remodeled(row):
    if pd.isna(row["Year Remod/Add"]):
        return "No Remodeled"
    elif row["Year Remod/Add"] > row["Year Built"]:
        return "New"
    else:
        return "Old"

train_dataset["new_IsHouseRemodeled"] = train_dataset.apply(check_house_remodeled, axis=1)
test_dataset["new_IsHouseRemodeled"] = test_dataset.apply(check_house_remodeled, axis=1)

print(train_dataset["new_IsHouseRemodeled"].value_counts())
```

```python
def check_exterior(row):
    if row["Exterior 1st"] != row["Exterior 2nd"]:
        return "yes"
    else:
        return "no"

train_dataset["new_has2Exterior"] = train_dataset.apply(check_exterior, axis=1)
test_dataset["new_has2Exterior"] = test_dataset.apply(check_exterior, axis=1)
```

```python
def check_mas_vnr_type(row):
    if pd.isna(row["Mas Vnr Type"]):
        return "No"
    else:
        return "Yes"

train_dataset["new_MasVnrType"] = train_dataset.apply(check_mas_vnr_type, axis=1)
test_dataset["new_MasVnrType"] = test_dataset.apply(check_mas_vnr_type, axis=1)

print(train_dataset["new_MasVnrType"])
```

```python
train_dataset["new_TotalBsmtFinSF"]=train_dataset["BsmtFin SF 1"]+train_dataset["BsmtFin SF 2"]
train_dataset["new_HasSecondFinishedBsmt"] = (train_dataset["BsmtFin SF 2"] > 0).astype(int)

train_dataset["new_TotalFlrSF"] = train_dataset["1st Flr SF"] + train_dataset["2nd Flr SF"]
train_dataset["new_HasSecondFloor"] = (train_dataset["2nd Flr SF"] > 0).astype(int)

train_dataset["new_TotalBath"] = train_dataset["Full Bath"] + train_dataset["Half Bath"] + train_dataset["Bsmt Full Bath"] + train_dataset["Bsmt Half Bath"]

test_dataset["new_TotalBsmtFinSF"]=test_dataset["BsmtFin SF 1"]+test_dataset["BsmtFin SF 2"]
test_dataset["new_HasSecondFinishedBsmt"] = (test_dataset["BsmtFin SF 2"] > 0).astype(int)

test_dataset["new_TotalFlrSF"] = test_dataset["1st Flr SF"] + test_dataset["2nd Flr SF"]
test_dataset["new_HasSecondFloor"] = (test_dataset["2nd Flr SF"] > 0).astype(int)

test_dataset["new_TotalBath"] = test_dataset["Full Bath"] + test_dataset["Half Bath"] + test_dataset["Bsmt Full Bath"] + test_dataset["Bsmt Half Bath"]



```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor

remaining_object_columns_train = train_dataset.select_dtypes(
    include=["object"]
).columns.tolist()

if remaining_object_columns_train:
    print(
        "Applying one-hot encoding to remaining object columns:",
        remaining_object_columns_train
    )

    train_dataset = pd.get_dummies(
        train_dataset,
        columns=remaining_object_columns_train,
        drop_first=True,
        dtype=int
    )

    test_dataset = pd.get_dummies(
        test_dataset,
        columns=remaining_object_columns_train,
        drop_first=True,
        dtype=int
    )

X = train_dataset.drop("SalePrice", axis=1)
y = train_dataset["SalePrice"].copy()

test_dataset = test_dataset.reindex(
    columns=X.columns,
    fill_value=0
)
```

```python
models = {
    "Linear Regression": LinearRegression(),

    "Random Forest": RandomForestRegressor(
        n_estimators=300,
        random_state=42,
        n_jobs=-1
    ),

    "XGBoost": XGBRegressor(
        n_estimators=500,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        objective="reg:squarederror",
        random_state=42,
        n_jobs=-1
    ),

    "CatBoost": CatBoostRegressor(
        iterations=500,
        learning_rate=0.05,
        depth=6,
        loss_function="RMSE",
        random_seed=42,
        verbose=0
    ),

    "LightGBM": LGBMRegressor(
        n_estimators=500,
        learning_rate=0.05,
        num_leaves=31,
        max_depth=-1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1,
        verbosity=-1
    )
}
```

```python
from sklearn.impute import SimpleImputer # Import SimpleImputer
from sklearn.pipeline import Pipeline # Import Pipeline

kfold = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
results = []

# Drop rows where 'y' (SalePrice) is NaN AFTER log transformation
# This ensures that cross_val_score receives a clean target variable.
valid_indices = y.dropna().index
X_cleaned = X.loc[valid_indices]
y_cleaned = y.loc[valid_indices]

for model_name, model in models.items():
    # Create a pipeline with imputation and the current model
    model_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='mean')),  # Impute missing numerical values in X
        ('regressor', model) # The actual regressor model
    ])

    rmse_scores = -cross_val_score(
        estimator=model_pipeline, # Use the pipeline here
        X=X_cleaned,
        y=y_cleaned,
        cv=kfold,
        scoring="neg_root_mean_squared_error",
        n_jobs=-1
    )

    r2_scores = cross_val_score(
        estimator=model_pipeline, # Use the pipeline here
        X=X_cleaned,
        y=y_cleaned,
        cv=kfold,
        scoring="r2",
        n_jobs=-1
    )

    results.append({
        "Model": model_name,
        "Mean RMSE": rmse_scores.mean(),
        "Std RMSE": rmse_scores.std(),
        "Mean R2": r2_scores.mean(),
        "Std R2": r2_scores.std()
    })

    print(f"{model_name} completed.")
```

```python
results_df = pd.DataFrame(results)

results_df = results_df.sort_values(
    by="Mean RMSE",
    ascending=True
).reset_index(drop=True)

print(results_df)
```

```python
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import (
    KFold,
    RandomizedSearchCV,
    GridSearchCV
)

from catboost import CatBoostRegressor


# =========================================================
# 1. X ve y'yi oluştur
# =========================================================

X = train_dataset.drop(columns=["SalePrice"])
y = train_dataset["SalePrice"]


# =========================================================
# 2. Sütun tiplerini belirle
# =========================================================

numeric_columns = X.select_dtypes(
    include=[np.number, "bool"]
).columns.tolist()

categorical_columns = X.select_dtypes(
    exclude=[np.number, "bool"]
).columns.tolist()

print("Numerical columns:", len(numeric_columns))
print("Categorical columns:", len(categorical_columns))


# =========================================================
# 3. Preprocessing
# CatBoost için StandardScaler gerekli değil
# =========================================================

numeric_pipeline = Pipeline([
    (
        "imputer",
        SimpleImputer(strategy="median")
    )
])

categorical_pipeline = Pipeline([
    (
        "imputer",
        SimpleImputer(strategy="most_frequent")
    ),
    (
        "onehot",
        OneHotEncoder(
            handle_unknown="ignore",
            sparse_output=True
        )
    )
])

preprocessor = ColumnTransformer(
    transformers=[
        (
            "num",
            numeric_pipeline,
            numeric_columns
        ),
        (
            "cat",
            categorical_pipeline,
            categorical_columns
        )
    ],
    remainder="drop"
)


# =========================================================
# 4. CatBoost pipeline
# =========================================================

catboost_pipeline = Pipeline([
    (
        "preprocessor",
        preprocessor
    ),
    (
        "regressor",
        CatBoostRegressor(
            loss_function="RMSE",
            random_seed=42,
            verbose=0,
            thread_count=1,
            allow_writing_files=False
        )
    )
])


# =========================================================
# 5. Cross-validation
# =========================================================

optimization_kfold = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)


# =========================================================
# 6. Randomized Search
# =========================================================

catboost_random_params = {
    "regressor__iterations": [
        300,
        500,
        700,
        1000,
        1300
    ],
    "regressor__depth": [
        4,
        5,
        6,
        7,
        8,
        9
    ],
    "regressor__learning_rate": [
        0.01,
        0.03,
        0.05,
        0.08,
        0.1
    ],
    "regressor__l2_leaf_reg": [
        1,
        3,
        5,
        7,
        10
    ],
    "regressor__random_strength": [
        0,
        0.5,
        1,
        2,
        3
    ]
}

catboost_random_search = RandomizedSearchCV(
    estimator=catboost_pipeline,
    param_distributions=catboost_random_params,
    n_iter=25,
    scoring="neg_root_mean_squared_error",
    cv=optimization_kfold,
    random_state=42,
    n_jobs=-1,
    verbose=2,
    return_train_score=True,
    refit=True
)

catboost_random_search.fit(X, y)

print("\nRandomized Search best parameters:")
print(catboost_random_search.best_params_)

print(
    "Randomized Search best CV RMSE:",
    -catboost_random_search.best_score_
)
```

```python
# Randomized Search sonuçlarını sakla
catboost_random_best_params = (
    catboost_random_search.best_params_
)

catboost_random_cv_rmse = (
    -catboost_random_search.best_score_
)

catboost_random_best_model = (
    catboost_random_search.best_estimator_
)

catboost_random_results = {
    "Model": "CatBoost",
    "Search": "RandomizedSearchCV",
    "Best CV RMSE": catboost_random_cv_rmse,
    "Best Parameters": catboost_random_best_params
}

print("\nCatBoost Randomized Search Results:")
print(catboost_random_results)
```

```python
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from xgboost import XGBRegressor


# =========================================================
# 1. Sütunları ayır
# =========================================================

numeric_columns = X.select_dtypes(
    include=[np.number, "bool"]
).columns.tolist()

categorical_columns = X.select_dtypes(
    exclude=[np.number, "bool"]
).columns.tolist()

print("Numerical columns:", len(numeric_columns))
print("Categorical columns:", len(categorical_columns))


# =========================================================
# 2. Preprocessing
# XGBoost için StandardScaler gerekli değil
# =========================================================

numeric_pipeline = Pipeline([
    (
        "imputer",
        SimpleImputer(strategy="median")
    )
])

categorical_pipeline = Pipeline([
    (
        "imputer",
        SimpleImputer(strategy="most_frequent")
    ),
    (
        "onehot",
        OneHotEncoder(
            handle_unknown="ignore",
            sparse_output=True
        )
    )
])

preprocessor = ColumnTransformer([
    (
        "num",
        numeric_pipeline,
        numeric_columns
    ),
    (
        "cat",
        categorical_pipeline,
        categorical_columns
    )
], remainder="drop")


# =========================================================
# 3. XGBoost pipeline
# =========================================================

xgboost_pipeline = Pipeline([
    (
        "preprocessor",
        preprocessor
    ),
    (
        "regressor",
        XGBRegressor(
            objective="reg:squarederror",
            random_state=42,
            n_jobs=1,
            verbosity=0
        )
    )
])


# =========================================================
# 4. Randomized Search
# =========================================================

xgboost_random_params = {
    "regressor__n_estimators": [
        300,
        500,
        700,
        1000,
        1300
    ],
    "regressor__learning_rate": [
        0.01,
        0.03,
        0.05,
        0.08,
        0.1
    ],
    "regressor__max_depth": [
        2,
        3,
        4,
        5,
        6,
        7
    ],
    "regressor__min_child_weight": [
        1,
        3,
        5,
        7,
        10
    ],
    "regressor__subsample": [
        0.6,
        0.7,
        0.8,
        0.9,
        1.0
    ],
    "regressor__colsample_bytree": [
        0.6,
        0.7,
        0.8,
        0.9,
        1.0
    ],
    "regressor__reg_alpha": [
        0,
        0.001,
        0.01,
        0.1,
        1
    ],
    "regressor__reg_lambda": [
        0.5,
        1,
        2,
        5,
        10
    ]
}

xgboost_random_search = RandomizedSearchCV(
    estimator=xgboost_pipeline,
    param_distributions=xgboost_random_params,
    n_iter=30,
    scoring="neg_root_mean_squared_error",
    cv=optimization_kfold,
    random_state=42,
    n_jobs=-1,
    verbose=2,
    return_train_score=True,
    refit=True
)

xgboost_random_search.fit(X, y)

xgboost_random_rmse = -xgboost_random_search.best_score_

print("\nRandomized Search best parameters:")
print(xgboost_random_search.best_params_)

print(
    f"Randomized Search best CV RMSE: "
    f"{xgboost_random_rmse:.6f}"
)
```

```python
# =========================================================
# 5. Randomized Search sonuçlarını sakla
# =========================================================

best_xgb_index = xgboost_random_search.best_index_

xgboost_random_best_params = (
    xgboost_random_search.best_params_
)

xgboost_random_best_model = (
    xgboost_random_search.best_estimator_
)

# En iyi adayın cross-validation train RMSE değeri
xgboost_random_train_rmse = -(
    xgboost_random_search.cv_results_[
        "mean_train_score"
    ][best_xgb_index]
)

# En iyi adayın cross-validation validation RMSE değeri
xgboost_random_cv_rmse = -(
    xgboost_random_search.best_score_
)

# Fold sonuçlarının standart sapması
xgboost_random_cv_std = (
    xgboost_random_search.cv_results_[
        "std_test_score"
    ][best_xgb_index]
)

# Train-validation farkı
xgboost_random_cv_gap = (
    xgboost_random_cv_rmse
    - xgboost_random_train_rmse
)

print("\nXGBoost Randomized Search Results")
print("-" * 45)

print(
    f"Train CV RMSE: "
    f"{xgboost_random_train_rmse:.6f}"
)

print(
    f"Validation CV RMSE: "
    f"{xgboost_random_cv_rmse:.6f}"
)

print(
    f"CV RMSE Standard Deviation: "
    f"{xgboost_random_cv_std:.6f}"
)

print(
    f"Train-Validation Gap: "
    f"{xgboost_random_cv_gap:.6f}"
)

print("\nBest Parameters:")
print(xgboost_random_best_params)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def plot_pipeline_feature_importance(
    fitted_pipeline,
    model_name,
    top_n=20
):
    """
    Fitted sklearn pipeline içindeki preprocessor ve regressor
    adımlarını kullanarak feature importance grafiği oluşturur.
    """

    preprocessor = fitted_pipeline.named_steps["preprocessor"]
    model = fitted_pipeline.named_steps["regressor"]

    # One-hot encoding sonrasında oluşan gerçek özellik isimleri
    feature_names = preprocessor.get_feature_names_out()

    # Modelin importance değerleri
    importance_values = model.feature_importances_

    # Özellik sayısı ile importance sayısını kontrol et
    if len(feature_names) != len(importance_values):
        raise ValueError(
            f"Feature name count ({len(feature_names)}) "
            f"does not match importance count "
            f"({len(importance_values)})."
        )

    feature_importance_df = pd.DataFrame({
        "Feature": feature_names,
        "Importance": importance_values
    })

    # ColumnTransformer ön eklerini temizle:
    # num__Overall Qual → Overall Qual
    # cat__Neighborhood_NAmes → Neighborhood_NAmes
    feature_importance_df["Feature"] = (
        feature_importance_df["Feature"]
        .str.replace(r"^num__", "", regex=True)
        .str.replace(r"^cat__", "", regex=True)
    )

    feature_importance_df = (
        feature_importance_df
        .sort_values(
            by="Importance",
            ascending=False
        )
        .reset_index(drop=True)
    )

    top_features = feature_importance_df.head(top_n)

    plt.figure(figsize=(10, 8))

    plt.barh(
        top_features["Feature"],
        top_features["Importance"]
    )

    plt.gca().invert_yaxis()
    plt.title(
        f"Top {top_n} {model_name} Feature Importances"
    )
    plt.xlabel("Importance")
    plt.ylabel("Feature")
    plt.tight_layout()
    plt.show()

    return feature_importance_df
```

```python
import shap
import pandas as pd
from scipy import sparse


# =========================================================
# LightGBM SHAP analizi
# Randomized + Grid Search sonucundaki nihai model kullanılır
# =========================================================

best_lightgbm_pipeline = lightgbm_grid_best_model

lightgbm_preprocessor = (
    best_lightgbm_pipeline
    .named_steps["preprocessor"]
)

best_lightgbm_regressor = (
    best_lightgbm_pipeline
    .named_steps["regressor"]
)


# Önce orijinal X içerisinden örnek al
X_shap_original_sample = X.sample(
    n=min(500, len(X)),
    random_state=42
)


# Örnek veriyi fitted preprocessor ile dönüştür
X_lightgbm_transformed = (
    lightgbm_preprocessor
    .transform(X_shap_original_sample)
)


# OneHotEncoder sparse çıktı ürettiği için
# SHAP öncesinde dense array'e dönüştür
if sparse.issparse(X_lightgbm_transformed):
    X_lightgbm_transformed = (
        X_lightgbm_transformed.toarray()
    )


# Dönüşüm sonrasındaki sütun isimleri
lightgbm_feature_names = (
    lightgbm_preprocessor
    .get_feature_names_out()
)


# Sütun isimlerindeki num__ ve cat__ ön eklerini temizle
lightgbm_feature_names = np.array([
    feature_name
    .replace("num__", "", 1)
    .replace("cat__", "", 1)

    for feature_name in lightgbm_feature_names
])


# SHAP için DataFrame oluştur
X_lightgbm_shap = pd.DataFrame(
    X_lightgbm_transformed,
    columns=lightgbm_feature_names,
    index=X_shap_original_sample.index
)


print(
    "SHAP sample shape:",
    X_lightgbm_shap.shape
)

print(
    "Model feature count:",
    best_lightgbm_regressor.n_features_in_
)
```

```python
# =========================================================
# SHAP değerlerini hesapla
# =========================================================

lightgbm_explainer = shap.TreeExplainer(
    best_lightgbm_regressor
)

lightgbm_shap_values = lightgbm_explainer(
    X_lightgbm_shap
)
```

```python
# =========================================================
# SHAP beeswarm grafiği
# =========================================================

shap.plots.beeswarm(
    lightgbm_shap_values,
    max_display=20
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import cross_val_predict
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    r2_score
)


# =========================================================
# 1. Nihai LightGBM pipeline
# =========================================================

best_lightgbm_pipeline = lightgbm_grid_best_model


# =========================================================
# 2. Cross-validation tahminleri
# Her gözlem, eğitimde kullanılmadığı fold'da tahmin edilir
# =========================================================

lightgbm_cv_predictions = cross_val_predict(
    estimator=best_lightgbm_pipeline,
    X=X,
    y=y,
    cv=optimization_kfold,
    n_jobs=-1,
    method="predict"
)


# =========================================================
# 3. Residual değerlerini hesapla
# Residual = Gerçek değer - Tahmin
# =========================================================

lightgbm_cv_residuals = (
    y.to_numpy()
    - lightgbm_cv_predictions
)


# =========================================================
# 4. Cross-validation performans değerleri
# =========================================================

lightgbm_oof_rmse = np.sqrt(
    mean_squared_error(
        y,
        lightgbm_cv_predictions
    )
)

lightgbm_oof_mae = mean_absolute_error(
    y,
    lightgbm_cv_predictions
)

lightgbm_oof_r2 = r2_score(
    y,
    lightgbm_cv_predictions
)

print("LightGBM out-of-fold results")
print("-" * 40)
print(f"RMSE: {lightgbm_oof_rmse:.6f}")
print(f"MAE:  {lightgbm_oof_mae:.6f}")
print(f"R²:   {lightgbm_oof_r2:.6f}")
```

```python
# =========================================================
# 5. Residual grafiği
# =========================================================

plt.figure(figsize=(10, 7))

plt.scatter(
    lightgbm_cv_predictions,
    lightgbm_cv_residuals,
    alpha=0.6
)

plt.axhline(
    y=0,
    linestyle="--",
    linewidth=2
)

plt.title(
    "LightGBM Residual Plot\n"
    "Randomized Search + Grid Search"
)

plt.xlabel("Predicted SalePrice")
plt.ylabel("Residual (Actual - Predicted)")
plt.tight_layout()
plt.show()
```
